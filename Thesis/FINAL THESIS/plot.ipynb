```python
y =    [0,
        0.0699,
        0.1936,
        0.4223,
        0.8832,
        1.3224]
```

```python
x =  [0,
        0.0002, 
        0.0005,
        0.001,
        0.002,
        0.003]
```

```python
p = []
for i in range(len(x)):
    p.append([x[i],y[i]])
```

```python
plt.figure(figsize=(6,4))
print(p)
plt.plot(x,y,'r-')
plt.plot(x,y,'ro')
plt.xlabel('Initial Strain [mm]')
plt.ylabel('Global Expansion [%]')
plt.title("Global Expansion vs. Initial Strain in Each Step")
#plt.xticks(np.arange(0, 21, 2.0))
plt.savefig("ASRA30P75_exp.png",dpi = 200)
plt.show()
```
