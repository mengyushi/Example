```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

# General

## Linear Regression

```python
df = pd.read_excel('Simulation_Result.xlsx')

version = 13
df = df.loc[df['Ver']==version].loc[df['CS']>0].loc[df[0]>=0].loc[df['Exp Raw']>=0]

df = df.loc[df['SUM OF CRACK']>=0].loc[df['Agg']==30].loc[df['BD']==1]

reg = linear_model.LinearRegression()

crack_w = [0,0.00005,0.0001,0.0002,0.0005,0.001,0.003,0.01,0.03]

for w in crack_w:
    
    curr_x = np.array(df[w].tolist()).reshape(-1,1)
    
    if w == crack_w[0]:
        x = curr_x
        
        
    else:
        x = np.append(x,curr_x,axis = 1)
y = np.array(df['Lost CS'].tolist()).reshape(-1,1)

#x = np.dot(x,w)

reg.fit(x,y)

print(reg.coef_*10000000)

plt.scatter(reg.predict(x),y)
plt.plot([0,1],[0,1],'r--')
```

## LASSO

```python
df = pd.read_excel('Simulation_Result.xlsx')

version = 13
df = df.loc[df['Ver']==version].loc[df['CS']>0].loc[df[0]>=0].loc[df['Exp Raw']>=0]

df = df.loc[df['SUM OF CRACK']>=0].loc[df['Agg']==30].loc[df['BD']==1]

reg = linear_model.Lasso(alpha=0.1,positive = True)


crack_w = [0,0.00005,0.0001,0.0002,0.0005,0.001,0.003,0.01,0.03]

for w in crack_w:
    
    curr_x = np.array(df[w].tolist()).reshape(-1,1)
    
    if w == crack_w[0]:
        x = curr_x
        
        
    else:
        x = np.append(x,curr_x,axis = 1)
y = np.array(df['Lost CS'].tolist()).reshape(-1,1)

#x = np.dot(x,w)

reg.fit(x,y)

print(reg.coef_*10000000)

plt.scatter(reg.predict(x),y)
plt.plot([0,1],[0,1],'r--')
```
