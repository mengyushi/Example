```python
df13 = pd.read_excel('Simulation_Result.xlsx')
```

```python
version = 13
df13 = df13.loc[df13['Ver']==version]
A30P75 = df13.loc[df13['Y']==-1].loc[df13['Agg']==30].loc[df13['Per']==75].loc[df13['BD']==1]
A30P75 
```

```python
plt.figure(figsize=(4,4))


ax = df.loc[df['Type']=='ASR'].loc[df['Resource']=='Swamy & Al-Asali (1988) 1 Mix 4.5% (From ALKAN HAFCI (2013)*)'].plot.scatter(x = 'Expansion (%)', y = 'Compressive Strength Remained',color = '#c4f503',label = 'Swamy & Al-Asali (1988) 1 Mix 4.5%')

df.loc[df['Type']=='ASR'].loc[df['Resource']=='Swamy & Al-Asali (1988) 1 Mix 15% (From ALKAN HAFCI (2013)*)'].plot.scatter(x = 'Expansion (%)', y = 'Compressive Strength Remained', ax = ax ,color = '#2e00ff',label = 'Swamy & Al-Asali (1988) 1 Mix 15%')

df.loc[df['Type']=='ASR'].loc[df['Resource']=='Swamy & Al-Asali (1988) 2 Mix 4.5% (From ALKAN HAFCI (2013)*)'].plot.scatter(x = 'Expansion (%)', y = 'Compressive Strength Remained', ax = ax ,color = '#ff7b00',label = 'Swamy & Al-Asali (1988) 2 Mix 4.5%')

df.loc[df['Type']=='ASR'].loc[df['Resource']=='Swamy & Al-Asali (1988) 2 Mix 15% (From ALKAN HAFCI (2013)*)'].plot.scatter(x = 'Expansion (%)', y = 'Compressive Strength Remained', ax = ax ,color = '#00ff50',label = 'Swamy & Al-Asali (1988) 2 Mix 15%')

A30P75.plot.scatter(x = 'EXP %', y = 'Residual CS',color = '#ff0000',ax = ax)
A30P75.plot(x = 'EXP %', y = 'Residual CS',color = '#ff0000',ax = ax ,label = 'A30P75 Simulation Result')

##################### GRAPH SETTING #####################

plt.xlabel('EXPANSION',fontsize='large')
plt.ylabel("RESIDUAL CS",fontsize='large')
plt.title(  'Residual Strength vs. Expansion' ,fontsize = 'large')
plt.legend(fontsize = 'small',loc = 1)
plt.tight_layout()
plt.savefig('A30P75_CS',dpi = 1000)
plt.show()
```
