```python
from __future__ import absolute_import, division, print_function
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from keras import models
from keras import layers
from keras import optimizers
from keras.wrappers import scikit_learn


from sklearn.preprocessing import PolynomialFeatures  
from sklearn.linear_model import LinearRegression

from sklearn.externals import joblib

import pickle

from keras.models import load_model

seed = 1
np.random.seed(seed)
```

```python
train_x = exp_df.loc[exp_df['Type']=='ASR'].loc[exp_df['Damaged Compressive Strength']>0]['Expansion (%)'].as_matrix()
train_y = exp_df.loc[exp_df['Type']=='ASR'].loc[exp_df['Damaged Compressive Strength']>0]['Compressive Strength Loss'].as_matrix()
train_x = train_x.reshape(train_x.shape[0],1)
```

```python
poly_features = PolynomialFeatures(degree = 5)  
train_data_poly = poly_features.fit_transform(train_x)
#test_data_poly = poly_features.fit_transform(test_x)
ASR_CS_regr = LinearRegression()
ASR_CS_regr.fit(train_data_poly, train_y)
train_y_pred = ASR_CS_regr.predict(train_data_poly)
#print('Coefficients: \n', regr.coef_)

plt.scatter(train_y, train_y_pred)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.xlim(plt.xlim())
plt.ylim(plt.ylim())
_ = plt.plot([-1, 1], [-1, 1],'r--')
    
print("")
print('Mean Abs Error: ',abs((train_y_pred-train_y)/len(train_y)).sum())
```

```python
def single_pred(regr_model,poly_features, exp):
    try:
        x = np.array([exp]).reshape(1, -1)
        x_poly = poly_features.fit_transform(x)
        train_y_pred = regr_model.predict(x_poly)
        return train_y_pred[0]
    except:
        return 'NaN'
```
