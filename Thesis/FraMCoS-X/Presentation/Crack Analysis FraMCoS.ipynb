```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def type():
    if Y == -1:
        return 'ASR'
    elif Y == 1:
        return 'DEF'
    else:
        return 'ASR+DEF'
    
df = pd.read_excel('Simulation_Result.xlsx')

version = 13
df = df.loc[df['Ver']==version].loc[df['CS']>0].loc[df['Exp Raw']>0]
```

```python
#df = df.loc[df['SUM OF CRACK']>=0].loc[df['Agg']==30].loc[df['BD']==1]
df = df.loc[df['BD']==1]
```

```python
width = 0.01

x = width/10
y = 'Residual CS' 

colors = ['#ff0000','#ff00ff','#ffbf00','#0000ff',
          '#00ffff','#33cc00']

###########################

ASR_A30_P100_num = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==100][x].tolist()
ASR_A30_P100_cs = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==100][y].tolist()

ASR_A30_P75_num = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==75][x].tolist()
ASR_A30_P75_cs = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==75][y].tolist()

ASR_A30_P25_num = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==25][x].tolist()
ASR_A30_P25_cs = df.loc[df['Type']=='ASR'].loc[df['Agg']==30].loc[df['Per']==25][y].tolist()

###########################

DEF_A30_I100_num = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==-1].loc[df['C']==1][x].tolist()
DEF_A30_I100_cs = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==-1].loc[df['C']==1][y].tolist()

DEF_A30_I75_num = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==-0.5].loc[df['C']==1][x].tolist()
DEF_A30_I75_cs = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==-0.5].loc[df['C']==1][y].tolist()

DEF_A30_I50_num = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==0].loc[df['C']==1][x].tolist()
DEF_A30_I50_cs = df.loc[df['Type']=='DEF'].loc[df['Agg']==30].loc[df['X']==0].loc[df['C']==1][y].tolist()

####################### PLOTTING #######################

fig, ax = plt.subplots(figsize=(4.2,4.2))
#plt.plot(exp_ASR_exp,exp_ASR_CS,'ko',mfc='none',label = "Experimental")

plt.plot(ASR_A30_P100_num,ASR_A30_P100_cs,'-',color = colors[0])
plt.plot(ASR_A30_P100_num,ASR_A30_P100_cs,'o',color = colors[0],label = 'ASRA30P100')

plt.plot(ASR_A30_P75_num,ASR_A30_P75_cs,'-',color = colors[1])
plt.plot(ASR_A30_P75_num,ASR_A30_P75_cs,'o',color = colors[1],label = 'ASRA30P75')

plt.plot(ASR_A30_P25_num,ASR_A30_P25_cs,'-',color = colors[2])
plt.plot(ASR_A30_P25_num,ASR_A30_P25_cs,'o',color = colors[2],label = 'ASRA30P25')


plt.plot(DEF_A30_I100_num,DEF_A30_I100_cs,'-',color = colors[3])
plt.plot(DEF_A30_I100_num,DEF_A30_I100_cs,'^',color = colors[3],label = 'DEFA30I100')

plt.plot(DEF_A30_I75_num,DEF_A30_I75_cs,'-',color = colors[4])
plt.plot(DEF_A30_I75_num,DEF_A30_I75_cs,'^',color = colors[4],label = 'DEFA30I75')

plt.plot(DEF_A30_I50_num,DEF_A30_I50_cs,'-',color = colors[5])
plt.plot(DEF_A30_I50_num,DEF_A30_I50_cs,'^',color = colors[5],label = 'DEFA30I50')

##################### GRAPH SETTING #####################
plt.xlim(-0.05,1.5)
plt.ylim(-0.1,1.2)
plt.xlabel("Total Cracked Interfaces \n Larger Than "+str(width)+" mm",fontsize = 'large')
plt.ylabel("Residual Compressive Strength",fontsize = 'large')
plt.title("Residual Compressive Strength \n vs. "+str(width)+"+mm Cracks",fontsize = 'large')
plt.legend(ncol = 1, loc=(1.02,0))
plt.tight_layout()
plt.autoscale()
plt.savefig("FINAL"+str(width)+".png",dpi = 1000)
plt.show()
```
