# PIXEL
----

    A small Program which can change JPG pic to pixel style.
    
    Could be run directly by PixelMaker.py

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import skimage
from skimage import data
import math
```

```python
im = plt.imread("test1.jpg")
plt.imshow(im)
```

```python
def pixel_maker(filename,resolution):
    
    im = plt.imread(filename)
    X = im.shape[0]
    Y = im.shape[1]
    
    if resolution > X or resolution > Y:
        print("Resolution required larger than origin.")
        return 
    

    piX = resolution
    piY = math.floor((piX/X)*Y)
    
    sec = math.floor(X/piX)
    
    small = np.zeros((piX,piY,3))
    
    for x in range(piX):
        for y in range(piY):
            small[x][y] = im[x*sec][y*sec]/256
    plt.axis('off')
    plt.imshow(small)
    plt.savefig("PIXEL_"+filename)
```

```python
pixel_maker("test1.jpg",10)
```
